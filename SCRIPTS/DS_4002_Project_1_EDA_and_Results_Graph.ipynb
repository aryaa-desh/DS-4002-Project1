```python


import json
import pandas as pd

file_path = '/Software_5 (2).json'

with open(file_path, 'r', encoding='utf-8') as json_file:
    data = [json.loads(line) for line in json_file]

df = pd.DataFrame(data)

display(df.head())
```

**Exploratory Data Analysis**







```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['overall'], bins=5, edgecolor='black')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Ratings')
plt.show()
```

Number of reviews over time

```python
import pandas as pd
import matplotlib.pyplot as plt

df['reviewTime'] = pd.to_datetime(df['reviewTime'])
reviews_over_time = df['reviewTime'].dt.year.value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(reviews_over_time.index, reviews_over_time.values)
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews Over Time (Yearly)')
plt.xticks(reviews_over_time.index.astype(int)) # Ensure years are whole numbers
plt.show()
```

```python
import pandas as pd

results_df = pd.read_csv('/results.csv')
display(results_df.head())
```

```python
# Calculate the average sentiment score per year
average_sentiment_per_year = results_df.groupby('year')['sentimentScore'].mean()

# Visualize the average sentiment per year
plt.figure(figsize=(12, 6))
plt.plot(average_sentiment_per_year.index, average_sentiment_per_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Sentiment Score')
plt.title('Average Sentiment Score Over Time')
plt.grid(True)
plt.xticks(average_sentiment_per_year.index.astype(int)) # Ensure years are whole numbers
plt.show()
```

```python
import matplotlib.pyplot as plt

average_sentiment_per_year = results_df.groupby('year')['sentimentScore'].mean()

sentiment_2000_2006 = average_sentiment_per_year[(average_sentiment_per_year.index >= 2000) & (average_sentiment_per_year.index <= 2006)]

plt.figure(figsize=(12, 6))
plt.plot(sentiment_2000_2006.index, sentiment_2000_2006.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Sentiment Score')
plt.title('Average Sentiment Score (2000-2006)')
plt.grid(True)
plt.xticks(sentiment_2000_2006.index.astype(int))
plt.show()
```
