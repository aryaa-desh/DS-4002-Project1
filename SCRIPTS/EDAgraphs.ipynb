```python
# imports 
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading in the json file of raw data row by row
raw_data = []
with open('Software_5.json', 'r') as file:
    for line in file:
        raw_data.append(json.loads(line))

df = pd.DataFrame(raw_data)

display(df.head())
```

```python
# plotting a histogram of rating distributions 
plt.figure(figsize=(8, 6))
plt.hist(df['overall'], bins=5, edgecolor='black')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Overall Ratings')
plt.show()
```

```python
# getting the number of reviews per years 
df['reviewTime'] = pd.to_datetime(df['reviewTime'])
reviews_over_time = df['reviewTime'].dt.year.value_counts().sort_index()

# plotting the number of reviews over time as a bar graph
plt.figure(figsize=(12, 6))
plt.bar(reviews_over_time.index, reviews_over_time.values)
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews Over Time (Yearly)')
plt.xticks(reviews_over_time.index.astype(int)) # Ensure years are whole numbers
plt.show()
```
